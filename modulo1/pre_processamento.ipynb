* **Finalmente, tratando os valores NaN**

```python
from sklearn.preprocessing.imputation import Imputer
imputer = Imputer(missing_values='NaN', strategy='mean', axis=0)
#from missing_values, use mean value as strategy
imputer = imputer.fit(previsores[:,0:3])
previsores[:,0:3] = imputer.transform(previsores[:,0:3])
```

# Escalonamento de Atributos
* O escalonamento é necessário quando há muita grande diferença na ordem dos valores(escala), de acordo com a coluna. Quando usar algoritmos euclidianos com essa diferença, o algoritmo considera a coluna com maior escala um atributo com muito maior importância, de acordo com a diferença de escala entre eles.

* O escalonamento faz com que os atributos possam ser interpretados como valores da mesma escala.

* Mesmo que não haja este erro de interpretação em função da escala, o escalonamento permite com que o algoritimo seja executado mais rapidamente.

## Maneiras de executar o escalonamento

* **Padronização(Standardisation)**

    `x = (x - v.média(x)) / desvio_padrão(x)` 
    
* **Normalização(Normalization)**
    
    `x = (x - v.mínimo(x) )/ (v.máximo(x) - v.mínimo(x))`

```python
from sklearn.preprocessing import StandardScaler
escala = StandardScaler()
escala.fit_transform(previsores)
```
